Label Ecnoding: Converting the label or categorical(text) into numeric form

```python
# importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

 Label Encoding of Breast Cancer Dataset


```python
# loading the data from csv file to pandas dataframe
cancer_data = pd.read_csv("/content/data.csv")
```

```python
cancer_data.head()
```

```python
# finding the count of different labels
cancer_data["diagnosis"].value_counts()
```

```python
# load the lable encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)  # learn categories and convert into numbers
```

```python
# append the labels to dataframe
cancer_data["target"] = labels
```

```python
cancer_data.head()
```

0--> Benign
1--> Malignant

```python
cancer_data["target"].value_counts()
```

Label Encoding of Iris Dataset
